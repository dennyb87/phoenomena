```python
import pandas as pd

period = [t/10 for t in range(8, 54)]
frequency = [1/t for t in period]

df = pd.DataFrame({"f": frequency}, index=period)
subdf = df[df.index * 10 % 10 == 0]
df.plot.line(grid=True, xticks=subdf.index, yticks=subdf.f.tolist(), xlabel="T (s)", ylabel="f (Hz)", figsize=[3, 3], legend=False)
```

```python
df = pd.DataFrame({"y": [25, 10, 5, 2, 1]}, index=[2, 5, 10, 25, 50])
df.plot.line(grid=True, xticks=df.index, yticks=df.y, xlabel="x", ylabel="y", figsize=[3, 2], legend=False)
```

```python
import math


lengths = [l/10 for l in range(11)]
t = [2*math.pi*math.sqrt(l/9.81) for l in lengths]
t_squared = [t**2 for t in t]

df = pd.DataFrame({"T": t, "$T^2$": t_squared}, index=lengths)
df.plot.line(grid=True, xticks=df.index, yticks=df["$T^2$"].tolist(), xlabel="L (m)", ylabel="T (s)", figsize=[3, 3], legend=True)
```
