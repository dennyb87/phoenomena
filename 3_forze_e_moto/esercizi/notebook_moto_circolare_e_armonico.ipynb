```python
import pandas as pd

period = [t/10 for t in range(8, 54)]
frequency = [1/t for t in period]

df = pd.DataFrame({"f": frequency}, index=period)
subdf = df[df.index * 10 % 10 == 0]
df.plot.line(grid=True, xticks=subdf.index, yticks=subdf.f.tolist(), xlabel="T (s)", ylabel="f (Hz)", figsize=[3, 3], legend=False)
```
