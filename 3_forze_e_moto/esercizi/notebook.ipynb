```python
import pandas as pd

speed = 100 # km/h
hours = [1/4, 1/2, 3/4, 1]

df = pd.DataFrame({"Treno": [speed * h for h in hours]}, index=hours)
df.plot.line(grid=True, xticks=df.index, yticks=df["Treno"], xlabel="t (h)", ylabel="s (km)", figsize=[4, 3])
```

```python
intercity_speed = 100 # km/h
regionale_speed = 50 # km/h

hours = [0, 1/4, 1/2, 3/4, 1]

intercity_distance = [intercity_speed * h for h in hours]
regionale_distance = [regionale_speed * h for h in hours]


df = pd.DataFrame({"intercity": intercity_distance, "regionale": regionale_distance}, index=hours)
df.plot.line(grid=True, xticks=df.index, yticks=df["intercity"], xlabel="t (h)", ylabel="s (km)", figsize=[4, 3])
```

```python
hours = [0, 1/4, 1/2, 3/4, 1]

speed = 0
initial_position = 50 # km/h


df = pd.DataFrame({"corpo statico": [initial_position + speed * h for h in hours]}, index=hours)
df.plot.line(grid=True, xticks=df.index, yticks=[100 * h for h in hours], xlabel="t (h)", ylabel="s (km)", figsize=[4, 3])
```

```python
speed_a = 80 # m/s
speed_b = 50 # m/s
times = [0, 1, 2, 3, 4]

df = pd.DataFrame({"a": [speed_a * t for t in times], "b": [speed_b * t for t in times]}, index=times)
df.plot.line(grid=True, xticks=df.index, yticks=df.a, xlabel="t (s)", ylabel="s (m)", figsize=[3, 2], legend=False)
```

```python
speed = 7.5 # m/s
seconds = [0, 3, 6, 9, 12]

df = pd.DataFrame({"y": [25 + (speed * s) for s in seconds]}, index=seconds)
df.plot.line(grid=True, xticks=df.index, yticks=pd.concat([pd.Series([0]), df.y]), xlabel="t (s)", ylabel="s (m)", figsize=[3, 2], legend=False)
```

```python
speed = 2 # m/s
seconds = [0, 1, 2, 3, 4]

df = pd.DataFrame({"y": [3 + (speed * s) for s in seconds]}, index=seconds)
df.plot.line(grid=True, xticks=df.index, yticks=range(0, 15, 3), xlabel="t (s)", ylabel="s (m)", figsize=[3, 2], legend=False)
```

```python
stops = [0, 16, 39, 67, 108]
minutes = [0, 18, 41, 69, 98]


df = pd.DataFrame({"intercity": stops}, index=minutes)
df.plot.line(grid=True, xticks=df.index, yticks=df.intercity, xlabel="t (min)", ylabel="s (km)", figsize=[3, 2], legend=False)
```

```python
speed_a = 17.5 # m/s
speed_b = 12.5 # m/s
seconds = range(0, 30, 2)

df = pd.DataFrame({"auto": [-350 + speed_a * t for t in seconds], "camion": [-230+speed_b * t for t in seconds]}, index=seconds)
df.plot.line(grid=True, xticks=df.index, yticks=df.auto, xlabel="t (s)", ylabel="s (m)", figsize=[5, 4])
```

```python
acceleration = 6.18 # m/s2
seconds = [0, 1, 2, 3, 4]

df = pd.DataFrame({"y": [acceleration * s for s in seconds]}, index=seconds)
df.plot.area(grid=True, xticks=df.index, yticks=df.y, xlabel="t (s)", ylabel="v (m/s)", figsize=[3, 2], legend=False)
```

```python
acceleration = 6.18 # m/s2
deciseconds = [n/10 for n in range(41)]

df = pd.DataFrame({"y": [0.5 * acceleration * s**2 for s in deciseconds]}, index=deciseconds)
subsample = df[df.index * 10 % 10 == 0]
df.plot.line(grid=True, xticks=subsample.index, yticks=subsample.y, xlabel="t (s)", ylabel="s (m)", figsize=[3, 3], legend=False)
```

```python
a_1 = 0.5 # m/s^2
a_2 = 0.0 # m/s^2

speed_1 = [a_1 * t for t in range(5)]  
speed_2 = [speed_1[-1] + a_2 * t for t in range(5, 11)]

speed_serie = speed_1 + speed_2 


df = pd.DataFrame({"y": speed_1 + speed_2}, index=range(len(speed_serie)))
df.plot.line(grid=True, xticks=df.index, yticks=df.y, xlabel="t (s)", ylabel="v (m/s)", figsize=[3, 2], legend=False)
```

```python
a_1 = 2.5  # m/s^2
a_2 = 0.0  # m/s^2
a_3 = -3.75 # m/s^2

speed_1 = [a_1 * t for t in range(13)]  
speed_2 = [speed_1[-1] + a_2 * t for t in range(10)]
speed_3 = [speed_2[-1] + a_3 * t for t in range(1, 9)]

speed_serie = speed_1 + speed_2 + speed_3

df = pd.DataFrame({"y": speed_serie}, index=range(0, 31))
df.plot.line(grid=True, xticks=df.index[df.index%2==0], yticks=df.y[df.y%1.5==0], xlabel="t (s)", ylabel="v (m/s)", figsize=[4, 2], legend=False)
```

```python
import math

radii = range(1, 6)
areas = [math.pi*r**2 for r in radii]

df = pd.DataFrame({"area": areas}, index=radii)
df.plot.line(grid=True, xticks=df.index, yticks=df.area, xlabel="radius (cm)", ylabel="area (cm\u00b2)", figsize=[3, 2], legend=False)
```
