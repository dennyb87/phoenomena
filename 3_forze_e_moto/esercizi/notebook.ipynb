```python
import pandas as pd

speed = 100 # km/h
hours = [1/4, 1/2, 3/4, 1]

df = pd.DataFrame({"Treno": [speed * h for h in hours]}, index=hours)
df.plot.line(grid=True, xticks=df.index, yticks=df["Treno"], xlabel="t (h)", ylabel="s (km)", figsize=[4, 3])
```

```python
intercity_speed = 100 # km/h
regionale_speed = 50 # km/h

hours = [0, 1/4, 1/2, 3/4, 1]

intercity_distance = [intercity_speed * h for h in hours]
regionale_distance = [regionale_speed * h for h in hours]


df = pd.DataFrame({"intercity": intercity_distance, "regionale": regionale_distance}, index=hours)
df.plot.line(grid=True, xticks=df.index, yticks=df["intercity"], xlabel="t (h)", ylabel="s (km)", figsize=[4, 3])
```

```python
hours = [0, 1/4, 1/2, 3/4, 1]

speed = 0
initial_position = 50 # km/h


df = pd.DataFrame({"corpo statico": [initial_position + speed * h for h in hours]}, index=hours)
df.plot.line(grid=True, xticks=df.index, yticks=[100 * h for h in hours], xlabel="t (h)", ylabel="s (km)", figsize=[4, 3])
```
