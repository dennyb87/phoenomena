```python
import pandas as pd
import math


mass_1 = 1 # kg
mass_2 = 2 # kg
forces = range(6)

a_1 = [force/mass_1 for force in forces]
a_2 = [force/mass_2 for force in forces]

df = pd.DataFrame({"1kg mass": a_1, "2kg mass": a_2}, index=forces)
df.plot.line(grid=True, xticks=df.index, yticks=df["1kg mass"], xlabel="F (N)", ylabel="a ($m/s^2$)", figsize=[3, 2], legend=True)
```

```python
force = 2 # N

masses = [m/10 for m in range(1, 35)]
acceleration = [force/mass for mass in masses]

df = pd.DataFrame({"y": acceleration}, index=masses)
subdf = df[df.index * 10 % 5 == 0]
df= df[df.y < 6]
ax = df.plot.line(grid=True, xticks=subdf.index, yticks=subdf.y, xlabel="m (kg)", ylabel="a ($m/s^2$)", figsize=[3, 3], legend=False)
ax.legend(["F = 2N"])
```

```python
mass = 250 # kg
force = range(50, 300, 50)
acceleration = [f/mass for f in force]

df = pd.DataFrame({"y": acceleration}, index=force)
ax = df.plot.line(grid=True, xticks=df.index, yticks=df.y, xlabel="F (N)", ylabel="a ($m/s^2$)", figsize=[3, 2], legend=True)
ax.legend(["mass 250kg"])
```
