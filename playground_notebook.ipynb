```python
import pandas as pd

df = pd.DataFrame({"Molla A": [0, 30, 60, 90], "Molla B": [0, 20, 40, 60]}, index=[0, 5, 10, 15])
df.plot.line(grid=True, xticks=df.index, yticks=df["Molla A"], xlabel="\u0394 L (cm)", ylabel="F (N)", figsize=[4, 3])
```

```python
g_earth = 9.8
g_moon = 1.6
mass_set = range(4)

df = pd.DataFrame({"Terra": [g_earth * m for m in mass_set], "Luna": [g_moon * m for m in mass_set]}, index=mass_set)
df.plot.line(grid=True, xticks=df.index, yticks=df["Terra"], xlabel="Massa (kg)", ylabel="Peso (N)", figsize=[4, 3])
```
